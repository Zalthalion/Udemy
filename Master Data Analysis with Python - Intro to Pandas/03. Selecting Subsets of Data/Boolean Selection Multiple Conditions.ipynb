```python
import pandas as pd
movies = pd.read_csv('../data/movie.csv', index_col='title')
movies.head()
```

## Exercises

### Exercise 1
<span  style="color:green; font-size:16px">Select all movies from the 1970s.</span>

```python
filter1 = movies['year'] >= 1970
filter2 = movies['year'] < 1980
unioned = filter1 & filter2
seventies = movies[unioned]
seventies.head()
```

### Exercise 2
<span  style="color:green; font-size:16px">Select all movies from the 1970s that had IMDB scores greater than 8.</span>

```python
filter3 = movies['imdb_score'] > 8
unioned  = filter1 & filter2 & filter3
good_seventies = movies[unioned]
good_seventies.head()
```

### Exercise 3
<span  style="color:green; font-size:16px">Select movies that were rated either R, PG-13, or PG.</span>

```python
rates = ['R', 'PG-13', 'PG']
filt = movies['content_rating'].isin(rates)
selected_rates = movies[filt]
selected_rates
```

### Exercise 4
<span  style="color:green; font-size:16px">Select movies that are either rated PG-13 or were made after 2010.</span>

```python
filter_1 = movies['content_rating'] == 'PG-13'
filter_2 = movies['year'] > 2010
filter_combined = filter_1 | filter_2
filtered_movies = movies[filter_combined]
filtered_movies
```

### Exercise 5
<span  style="color:green; font-size:16px">Find all the movies that have at least one of the three actors with more than 10,000 Facebook likes.</span>

```python
filt0 = movies['actor1_fb'] > 10000
filt2 = movies['actor2_fb'] > 10000
filt3 = movies['actor3_fb'] > 10000
filters_combined = filt0 | filt2 | filt3
popular_actors = movies[filters_combined]
popular_actors
```

### Exercise 6
<span  style="color:green; font-size:16px">Reverse the condition from problem 4. In words, what have you selected?</span>

```python
# None of the actors have 10k likes
unpopular_actors = movies[~filters_combined]
unpopular_actors
```
